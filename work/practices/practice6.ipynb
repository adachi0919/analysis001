```python
from bs4 import BeautifulSoup
import requests
import logging
import pandas as pd, numpy as np
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
url = "https://www.uta-net.com/renew/myutanet/index.php?mode=month"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
elems = soup.find_all("td", class_="artist_name")
artist_names = [elem.string for elem in elems]
artist_links = [elem.a["href"] for elem in elems]
music = np.array([artist_names, artist_links]).T
music = pd.DataFrame(music, columns=['アーティスト', 'リンク'])
```
