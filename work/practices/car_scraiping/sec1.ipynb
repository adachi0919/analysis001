```python
import requests
from bs4 import BeautifulSoup
import re
import uuid
```

```python
url = "https://www.google.com/search?q=honda+fit&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjM_uDC7bvtAhXMfXAKHaHiC44Q_AUoAnoECAwQBA&biw=834&bih=939"
res = requests.get(url)
res.status_code
```

```python
res.text
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
soup.find_all("img", src="^data:image")
```

```python
imgs = soup.find_all('img',src=re.compile('^https://msp.c.yimg.jp/yjimage'))
```

```python
soup.find_all('img', class_="rg_i Q4LuWd")
```

```python
print(soup.prettify())
```

```python
imgs = soup.find_all("img", src=re.compile("^https://"))
for img in imgs:
    r = requests.get(img['src'])
    with open(str('./picture/')+str(uuid.uuid4())+str('.jpeg'), 'wb') as f:
        f.write(r.content)
```
