```python
def MecabMorphologicalAnalysis(path, output_file='wakati.txt'):
    mecab_text = ''
    with open(path) as f:
        for line in f:
            mecab_text += mecab.parse(line)
    with open(output_file, 'w') as f:
        print(mecab_text, file=f)
```
