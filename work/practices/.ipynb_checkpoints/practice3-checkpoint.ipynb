```python
import logging
import pandas as pd
from gensim.models.word2vec import Word2Vec, Text8Corpus
from janome.tokenizer import Tokenizer
from math import log
```

```python
t = Tokenizer()
```

```python
df = pd.read_csv('/work/data/BunshuOnline/news.csv')
texts = df.news_page_list
```

```python
word_list = []
```

```python
for text in texts:
    wakati_text = list(t.tokenize(text, wakati=True))
    word_list.append(wakati_text)
```

```python
word_list
```

```python
unique_words = []
n = 0
for words in word_list:
    n += len(words)
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
```

```python
bow_list =[]
for words in word_list:
    bag =[]
    for unique_word in unique_words:
        num = words.count(unique_word)
        bag.append(num)
    bow_list.append(bag)
```

```python
idf = []
for i in range(len(unique_words)): # len(unique_words) => 全ての記事においてuniqueな単語数
    count = 0
    for bow in bow_list:
        if bow[i] > 0:
            count += 1 # 出現する言葉の種類数 => count
        idf.append(log((n+1)/(count+1))) # idf => 記事ごとのidf, n => 全ての記事の文字数合計
tfidfs = []
for bow in bow_list:
    tfidf = []
    n = sum(bow)
    for i, value in enumerate(bow):
        tf = value / n
        tfidf.append(tf * (idf[i] + 1))
    tfidfs.append(tfidf)
```
