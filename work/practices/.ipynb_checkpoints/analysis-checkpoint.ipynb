```python
import fasttext
import re
import numpy as np
from sklearn import preprocessing
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
model = fasttext.train_supervised("news.train", lr=0.5, epoch=100, wordNgrams=2, bucket=200000, dim=300, loss='ova')
```

```python
model.test("news.valid")
```

```python
with open("news.valid") as f:
    l_strip = [s.strip() for s in f.readlines()] # 改行文字除去
```

```python
l_strip
```

```python
s = l_strip[1]
print(s)
```

```python
re.findall('__label__(.*?) ', s)[0]
```

```python
labels = []
texts = []
sizes = []
for t in l_strip:
    if model.predict(re.findall(' (.*)', t))[1][0] > 0.0:
        labels.append(re.findall('__label__(.*?) ', t)[0])
        texts.append(re.findall(' (.*)', t)[0])
        sizes.append(model.predict(re.findall(' (.*)', t))[1][0][0])
```

```python
labels
```

```python
texts
```

```python
sizes
```

```python
vectors = []
for t in texts:
    vectors.append(model.get_sentence_vector(t))
```

```python
vectors = np.array(vectors)
```

```python
vectors.shape
```

```python
vectors
```

```python
labels = np.array(labels)
```

```python
labels
```

```python
sizes = np.array(sizes)
```

```python
sizes
```

```python
ss = preprocessing.StandardScaler()
```

```python
vectors_std = ss.fit_transform(vectors)
```

```python
vectors_std
```

```python
pca = PCA()
pca.fit(vectors_std)
feature = pca.transform(vectors_std)
feature = feature[:, :2]
```

```python
feature.shape
```

```python
feature
```

```python
labels
```

```python
np.unique(labels)
```

```python
x0, y0, z0 = feature[labels=='エンタメ', 0], feature[labels=='エンタメ', 1], sizes[labels=='エンタメ']**2*220
x1, y1, z1 = feature[labels=='スポーツ', 0], feature[labels=='スポーツ', 1], sizes[labels=='スポーツ']**2*220
x2, y2, z2 = feature[labels=='ライフ', 0], feature[labels=='ライフ', 1], sizes[labels=='ライフ']**2*220
x3, y3, z3 = feature[labels=='国内', 0], feature[labels=='国内', 1], sizes[labels=='国内']**2*220
x4, y4, z4 = feature[labels=='国際', 0], feature[labels=='国際', 1], sizes[labels=='国際']**2*220
x5, y5, z5 = feature[labels=='地域', 0], feature[labels=='地域', 1], sizes[labels=='地域']**2*220
x6, y6, z6 = feature[labels=='経済', 0], feature[labels=='経済', 1], sizes[labels=='経済']**2*220
```

```python
plt.title("Yahooニュース")
plt.scatter(x0, y0, label="エンタメ", s=z0)
plt.scatter(x1, y1, label="スポーツ", s=z1)
plt.scatter(x2, y2, label="ライフ", s=z2)
plt.scatter(x3, y3, label="国内", s=z3)
plt.scatter(x4, y4, label="国際", s=z4)
plt.scatter(x5, y5, label="地域", s=z5)
plt.scatter(x6, y6, label="経済", s=z6)
plt.legend(title="category")
plt.show()
```
