```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

data, label = make_moons(noise=0.1, n_samples=2000)
```

```python
data
```

```python
label
```

```python
x0, y0 = data[label==0, 0], data[label==0, 1]
x1, y1 = data[label==1, 0], data[label==1, 1]
```

```python
plt.title("moon")
plt.scatter(x0, y0, label="label-A")
plt.scatter(x1, y1, label="label-B")
plt.xlabel("X-LABEL")
plt.ylabel("Y-LABEL")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as pat

fig = plt.figure(figsize = (5, 5))
ax = fig.add_subplot(111)
ax.set_title("RGBA", fontsize = 14)

c1 = pat.Circle(xy = (0.2, 0.8), radius = 0.1,
                label = "(1, 0, 0, 0.2)", color = (1, 0, 0, 0.2))

c2 = pat.Circle(xy = (0.4, 0.6), radius = 0.1,
                label = "(1, 0, 0, 0.5)", color = (1, 0, 0, 0.5))

c3 = pat.Circle(xy = (0.6, 0.4), radius = 0.1,
                label = "(1, 0, 0, 1.0)", color = (1, 0, 0, 1.0))

ax.add_patch(c1)
ax.add_patch(c2)
ax.add_patch(c3)

ax.legend()
plt.show()
```
