```python
import pandas as pd, numpy as np
import re
import MeCab
import fasttext
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/work/data/BunshuOnline/yahoo_news2.csv')
```

```python
df
```

```python
cat_lst = []
for cat in df.category:
    cat_lst.append('__label__' + cat)
```

```python
cat_lst
```

```python
# MeCabの辞書にNEologdを指定。
# mecabは携帯素解析用、wakatiは分かち書き用
mecab = MeCab.Tagger('-d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd/')
wakati = MeCab.Tagger("-Owakati -d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd/")
```

```python
def MecabMorphologicalAnalysis(path='./text.txt', output_file='wakati.txt', mecab=mecab, file=False):
    mecab_text = ''
    if file:
        with open(path) as f:
            for line in f:
                mecab_text += mecab.parse(line)
        with open(output_file, 'w') as f:
            print(mecab_text, file=f)
    else:
        for path in path.split('\n'):
            mecab_text += mecab.parse(path)
        return mecab_text
```

```python
text = []
for tex in df.text:
    text.append(MecabMorphologicalAnalysis(tex, mecab=wakati))
```

```python
text
```

```python
df_train = pd.DataFrame({'cat': cat_lst, 'text': text})
```

```python
path = '/work/MorphologicalAnalysis/work/practices/train.txt'
```

```python
with open(path, mode='w') as f:
    for i in range(len(df_train)):
        f.write(df_train.iloc[i]['cat']+' '+df_train.iloc[i]['text'])
```
